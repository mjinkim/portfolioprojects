```python
# Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the conf of the plot

# Read in data

df = pd.read_csv(r'C:\Users\kimma\Downloads\movies.csv')

```

```python
#looking at data

df.head()
```

```python
# Missing Data ? 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types

df.dtypes
```

```python
# convert buget, gross, and votes values to remove decemials 

df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
df['votes'] = df['votes'].astype('Int64')
```

```python
# View by highest grossing film 

df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
# reviewing company to see any duplicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
#replaced NAN values in bugdet and gross with zero in order to use as int

df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)
```

```python
# Scatter plot = Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earning')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earning')

plt.show()
```

```python
# Plot Budget vs Gross Using seaborn 

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# viewing correlation 

df.corr(method='pearson')
```

```python
# Correlation Between Budget and Gross

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Feature')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Assigning a numerical value to objects within data set

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Feature')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
#unstacking correlation 

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
#Paired Correlation values 

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#Filter Correlation on values higher that 0.5 correlation 

high_corr = sorted_pairs[(sorted_pairs > 0.5)]

high_corr
```
